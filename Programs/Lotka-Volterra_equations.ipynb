```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw
```

```python

def derivative(X, t, alpha1, alpha2, a11,  a12,  a21,  a22):
    x, y = X
    dotx = x * (alpha1 + a11 * x + a12 * y)
    doty = y * (alpha2 + a21 * x + a22 * y)
    return np.array([dotx, doty])
```

# Primer ejemplo

```python
alpha1 = 4
alpha2 = -1
a11 = -0.0078
a12 = -0.05
a21 = 0.02
a22 = -0.01
x0 = 50
y0 = 6
Nt = 100000
tmax = 10
t = np.linspace(0.,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(derivative, X0, t, args = (alpha1, alpha2, a11,  a12,  a21,  a22))
x, y = res.T
```

```python
plt.figure()
plt.grid()
plt.title("Solución de la ecuación de Lotka-Voterra")
plt.plot(t, x,  label = 'Presas')
plt.plot(t, y, label = "Depredador")
plt.xlabel('Tiempo, [días]')
plt.ylabel('Población')
plt.legend()
plt.savefig('Solución_Experimento_1')
plt.show()
```

```python
plt.figure()
plt.grid()
plt.title("Diagrama de fases de Lotka-Volterra")
plt.plot(x, y)
plt.xlabel('Presas')
plt.ylabel('Depredadores')
plt.savefig('Diagrama_Experimento_1')
plt.show()
```
