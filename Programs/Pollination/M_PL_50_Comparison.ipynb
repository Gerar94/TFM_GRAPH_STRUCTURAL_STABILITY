Code to do the comparison in the Pollinization Netwroks 30x20

```python
#Libraries needed
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Read all the csv
df_PL_50 = pd.read_csv('Result_CSV/M_PL_50.csv', sep = ',', header=0) 
df_PL_EB = pd.read_csv('Result_CSV/PL_EB.csv', sep = ',', header=0)
df_PL_EC = pd.read_csv('Result_CSV/PL_EC.csv', sep = ',', header=0)
df_PL_ED = pd.read_csv('Result_CSV/PL_ED.csv', sep = ',', header=0)
```

# Comparison PL_50-EB

```python
#Comparison Mean degree=
plt.plot(df_PL_EB['meangrade'], df_PL_EB['Struc_Stab'], "b.",
         df_PL_50['meangrade'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Mean Grade')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison sum og gamma 
plt.plot(df_PL_EB['sum_coop'], df_PL_EB['Struc_Stab'], "b.", df_PL_50['sum_coop'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Sum Gamma')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison sum of beta
plt.plot(df_PL_EB['sum_comp'], df_PL_EB['Struc_Stab'], "b.", df_PL_50['sum_comp'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Sum Beta')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity eigenvector
plt.plot(df_PL_50['Modularity_eig'], df_PL_50['Struc_Stab'], "ro", df_PL_EB['Modularity_eig'], df_PL_EB['Struc_Stab'], "b.")
plt.xlabel('Modularity Eigenvector')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
#NOT POSSIBLE TO COMPUTE, TOO MUCH TIME
plt.plot(df_PL_EB['Modularity_opt'], df_PL_EB['Struc_Stab'], "b.",
         df_PL_50['Modularity_opt'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Modularity Optimal')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_PL_EB['Modularity_greedy'], df_PL_EB['Struc_Stab'], "b.",
         df_PL_50['Modularity_greedy'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Modularity Fast Greedy')
plt.ylabel('Structural Stability')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

# Spearman correlation
