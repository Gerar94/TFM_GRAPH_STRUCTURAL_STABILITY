Code to do the comparison in the Seed Dispersal Netwroks 20x20

```python
#Libraries needed
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Read all the csv
df_PL_50 = pd.read_csv('Result_CSV/M_PL_50.csv', sep = ',', header=0) 
df_PL_EB = pd.read_csv('Result_CSV/PL_EB.csv', sep = ',', header=0)
df_PL_EC = pd.read_csv('Result_CSV/PL_EC.csv', sep = ',', header=0)
df_PL_ED = pd.read_csv('Result_CSV/PL_ED.csv', sep = ',', header=0)
```

# Comparison PL_50-GB

```python
#Comparison Mean degree=
plt.plot(df_PL_EB['meangrade'], df_PL_EB['Struc_Stab'], "b.", label='M_PL_EB')
plt.plot(df_PL_50['meangrade'], df_PL_50['Struc_Stab'], "ro", label='M_PL_50')
plt.xlabel('Grado Medio')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_PL_50-PL_EB-Mean_Grade.png')
plt.show()
```

```python
#Comparison sum og gamma 
plt.plot(df_PL_EB['sum_coop'], df_PL_EB['Struc_Stab'], "b.",  label='M_PL_EB')
plt.plot(df_PL_50['sum_coop'], df_PL_50['Struc_Stab'], "ro",  label='M_PL_50')
plt.xlabel('Sum Gamma')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_PL_50-PL_EB-Sum_Gamma.png')
plt.show()
```

```python
#Comparison sum of beta
plt.plot(df_PL_EB['sum_comp'], df_PL_EB['Struc_Stab'], "b.", df_PL_50['sum_comp'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Sum Beta')
plt.ylabel('Estabilidad Estructural')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity eigenvector
plt.plot(df_PL_EB['Modularity_eig'], df_PL_EB['Struc_Stab'], "b.", label = 'M_PL_EB')
plt.plot(df_PL_50['Modularity_eig'], df_PL_50['Struc_Stab'], "ro", label = 'M_PL_50')
plt.xlabel('Modularidad Eigenvector')
plt.ylabel('Estabilidad Estructural')
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.grid(True)
plt.savefig('graphs/M_PL_50-PL_EB-ModulEigen.png')
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_PL_EB['Modularity_opt'], df_PL_EB['Struc_Stab'], "b.",
         df_PL_50['Modularity_opt'], df_PL_50['Struc_Stab'], "ro")
plt.xlabel('Modularity Óptimal')
plt.ylabel('Estabilidad Estructural')
plt.title('M_PL_50 - PL_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_PL_50['Modularity_opt'], df_PL_50['Struc_Stab'], "ro", label = 'M_PL_50')
plt.xlabel('Modularidad Óptima')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.grid(True)
plt.savefig('graphs/M_PL_50-ModulOpt.png')
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_PL_EB['Modularity_greedy'], df_PL_EB['Struc_Stab'], "b.", label = 'M_PL_EB')
plt.plot(df_PL_50['Modularity_greedy'], df_PL_50['Struc_Stab'], "ro", label = 'M_PL_50')
plt.xlabel('Modularidad Fast Greedy')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_PL_50-PL_EB-Modul.png')
plt.show()
```

# Spearman correlation

```python
#Spearman correlation M_PL_50
#SPearman correlation
spear_mean_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['meangrade'])
spear_coop_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['sum_coop'])
spear_comp_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['sum_comp'])
spear_modu_eig_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['Modularity_eig'])
spear_modu_opt_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['Modularity_opt'])
spear_modu_greedy_50 = stats.spearmanr(df_PL_50['Struc_Stab'], df_PL_50['Modularity_greedy'])
spear_mean_50.correlation, spear_coop_50.correlation ,spear_comp_50.correlation, spear_modu_eig_50.correlation, spear_modu_opt_50.correlation, spear_modu_greedy_50.correlation
```
