Code to do the comparison in the Seed Dispersal Netwroks 20x20

```python
#Libraries needed
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Read all the csv
df_SD_20 = pd.read_csv('Result_CSV/M_SD_20.csv', sep = ',', header=0) 
df_SD_GB = pd.read_csv('Result_CSV/SD_GB.csv', sep = ',', header=0)
df_SD_GC = pd.read_csv('Result_CSV/SD_GC.csv', sep = ',', header=0)
df_SD_GD = pd.read_csv('Result_CSV/SD_GD.csv', sep = ',', header=0)
```

# Comparison SD_20-GB

```python
#Comparison Mean degree=
plt.plot(df_SD_GB['meangrade'], df_SD_GB['Struc_Stab'], "b.",
         df_SD_20['meangrade'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Mean Grade')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison sum og gamma 
plt.plot(df_SD_GB['sum_coop'], df_SD_GB['Struc_Stab'], "b.", df_SD_20['sum_coop'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Sum Gamma')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison sum of beta
plt.plot(df_SD_GB['sum_comp'], df_SD_GB['Struc_Stab'], "b.", df_SD_20['sum_comp'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Sum Beta')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity eigenvector
plt.plot(df_SD_20['Modularity_eig'], df_SD_20['Struc_Stab'], "ro", df_SD_GB['Modularity_eig'], df_SD_GB['Struc_Stab'], "b.")
plt.xlabel('Modularity Eigenvector')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_GB['Modularity_opt'], df_SD_GB['Struc_Stab'], "b.",
         df_SD_20['Modularity_opt'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Modularity Optimal')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_GB['Modularity_greedy'], df_SD_GB['Struc_Stab'], "b.",
         df_SD_20['Modularity_greedy'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Modularity Fast Greedy')
plt.ylabel('Structural Stability')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

# Spearman correlation

```python
#Spearman correlation M_SD_20
#SPearman correlation
spear_mean_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['meangrade'])
spear_coop_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['sum_coop'])
spear_comp_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['sum_comp'])
spear_modu_eig_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_eig'])
spear_modu_opt_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_opt'])
spear_modu_greedy_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_greedy'])
spear_mean_20.correlation, spear_coop_20.correlation ,spear_comp_20.correlation, spear_modu_eig_20.correlation, spear_modu_opt_20.correlation, spear_modu_greedy_20.correlation
```
