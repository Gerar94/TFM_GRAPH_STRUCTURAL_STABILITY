Code to do the comparison in the Seed Dispersal Netwroks 20x20

```python
#Libraries needed
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Read all the csv
df_SD_20 = pd.read_csv('Result_CSV/M_SD_20.csv', sep = ',', header=0) 
df_SD_GB = pd.read_csv('Result_CSV/SD_GB.csv', sep = ',', header=0)
df_SD_GC = pd.read_csv('Result_CSV/SD_GC.csv', sep = ',', header=0)
df_SD_GD = pd.read_csv('Result_CSV/SD_GD.csv', sep = ',', header=0)
```

# Comparison SD_20-GB

```python
#Comparison Mean degree=
plt.plot(df_SD_GB['meangrade'], df_SD_GB['Struc_Stab'], "b.", label='M_SD_GB')
plt.plot(df_SD_20['meangrade'], df_SD_20['Struc_Stab'], "ro", label='M_SD_20')
plt.xlabel('Grado Medio')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_20-SD_GB-Mean_Grade.png')
plt.show()
```

```python
#Comparison sum og gamma 
plt.plot(df_SD_GB['sum_coop'], df_SD_GB['Struc_Stab'], "b.",  label='M_SD_GB')
plt.plot(df_SD_20['sum_coop'], df_SD_20['Struc_Stab'], "ro",  label='M_SD_20')
plt.xlabel('Sum Gamma')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_20-SD_GB-Sum_Gamma.png')
plt.show()
```

```python
#Comparison sum of beta
plt.plot(df_SD_GB['sum_comp'], df_SD_GB['Struc_Stab'], "b.", df_SD_20['sum_comp'], df_SD_20['Struc_Stab'], "ro")
plt.xlabel('Sum Beta')
plt.ylabel('Estabilidad Estructural')
plt.title('M_SD_20 - SD_GB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity eigenvector
plt.plot(df_SD_20['Modularity_eig'], df_SD_20['Struc_Stab'], "ro",  label='M_SD_20')
plt.plot(df_SD_GB['Modularity_eig'], df_SD_GB['Struc_Stab'], "b.", label='M_SD_GB')
plt.xlabel('Modularidad Eigenvector')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_20-SD_GB-ModulEigen.png')
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_GB['Modularity_opt'], df_SD_GB['Struc_Stab'], "b.", label='M_SD_GB')
plt.plot(df_SD_20['Modularity_opt'], df_SD_20['Struc_Stab'], "ro", label='M_SD_20')
plt.xlabel('Modularidad Óptima')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_20-SD_GB-ModulOpt.png')
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_GB['Modularity_greedy'], df_SD_GB['Struc_Stab'], "b.", label = 'M_SD_GB')
plt.plot(df_SD_20['Modularity_greedy'], df_SD_20['Struc_Stab'], "ro", label = 'M_SD_20')
plt.xlabel('Modularidad Fast Greedy')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_20-SD_GB-Modul.png')
plt.show()
```

# Spearman correlation

```python
#Spearman correlation M_SD_20
#SPearman correlation
spear_mean_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['meangrade'])
spear_coop_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['sum_coop'])
spear_comp_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['sum_comp'])
spear_modu_eig_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_eig'])
spear_modu_opt_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_opt'])
spear_modu_greedy_20 = stats.spearmanr(df_SD_20['Struc_Stab'], df_SD_20['Modularity_greedy'])
spear_mean_20.correlation, spear_coop_20.correlation ,spear_comp_20.correlation, spear_modu_eig_20.correlation, spear_modu_opt_20.correlation, spear_modu_greedy_20.correlation
```
