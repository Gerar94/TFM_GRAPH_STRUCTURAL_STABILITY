Code to do the comparison in the Seed Dispersal Netwroks 20x20

```python
#Libraries needed
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Read all the csv
df_SD_50 = pd.read_csv('Result_CSV/M_SD_50.csv', sep = ',', header=0) 
df_SD_EB = pd.read_csv('Result_CSV/SD_EB.csv', sep = ',', header=0)
df_SD_EC = pd.read_csv('Result_CSV/SD_EC.csv', sep = ',', header=0)
df_SD_ED = pd.read_csv('Result_CSV/SD_ED.csv', sep = ',', header=0)
```

# Comparison SD_50-GB

```python
#Comparison Mean degree=
plt.plot(df_SD_EB['meangrade'], df_SD_EB['Struc_Stab'], "b.", label='M_SD_EB')
plt.plot(df_SD_50['meangrade'], df_SD_50['Struc_Stab'], "ro", label='M_SD_50')
plt.xlabel('Grado Medio')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_50-SD_EB-Mean_Grade.png')
plt.show()
```

```python
#Comparison sum og gamma 
plt.plot(df_SD_EB['sum_coop'], df_SD_EB['Struc_Stab'], "b.",  label='M_SD_EB')
plt.plot(df_SD_50['sum_coop'], df_SD_50['Struc_Stab'], "ro",  label='M_SD_50')
plt.xlabel('Sum Gamma')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_50-SD_EB-Sum_Gamma.png')
plt.show()
```

```python
#Comparison sum of beta
plt.plot(df_SD_EB['sum_comp'], df_SD_EB['Struc_Stab'], "b.", df_SD_50['sum_comp'], df_SD_50['Struc_Stab'], "ro")
plt.xlabel('Sum Beta')
plt.ylabel('Estabilidad Estructural')
plt.title('M_SD_50 - SD_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity eigenvector
plt.plot(df_SD_EB['Modularity_eig'], df_SD_EB['Struc_Stab'], "b.", label = 'M_SD_EB')
plt.plot(df_SD_50['Modularity_eig'], df_SD_50['Struc_Stab'], "ro", label = 'M_SD_50')
plt.xlabel('Modularidad Eigenvector')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_50-SD_EB-ModulEigen.png')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_EB['Modularity_opt'], df_SD_EB['Struc_Stab'], "b.",
         df_SD_50['Modularity_opt'], df_SD_50['Struc_Stab'], "ro")
plt.xlabel('Modularity Óptimal')
plt.ylabel('Estabilidad Estructural')
plt.title('M_SD_50 - SD_EB')
plt.grid(True)
plt.show()
```

```python
#Comparison modularity opt
plt.plot(df_SD_EB['Modularity_greedy'], df_SD_EB['Struc_Stab'], "b.", label = 'M_SD_EB')
plt.plot(df_SD_50['Modularity_greedy'], df_SD_50['Struc_Stab'], "ro", label = 'M_SD_50')
plt.xlabel('Modularidad Fast Greedy')
plt.ylabel('Estabilidad Estructural')
plt.grid(True)
plt.legend(frameon=False, loc = 'upper center', ncol = 2, bbox_to_anchor =(0.5,1.1))
plt.savefig('graphs/M_SD_50-SD_EB-Modul.png')
plt.show()
```

# Spearman correlation

```python
#Spearman correlation M_SD_50
#SPearman correlation
spear_mean_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['meangrade'])
spear_coop_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['sum_coop'])
spear_comp_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['sum_comp'])
spear_modu_eig_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['Modularity_eig'])
spear_modu_opt_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['Modularity_opt'])
spear_modu_greedy_50 = stats.spearmanr(df_SD_50['Struc_Stab'], df_SD_50['Modularity_greedy'])
spear_mean_50.correlation, spear_coop_50.correlation ,spear_comp_50.correlation, spear_modu_eig_50.correlation, spear_modu_opt_50.correlation, spear_modu_greedy_50.correlation
```
