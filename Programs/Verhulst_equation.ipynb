```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f_logistica(p0,K,r,t):
    numerador = K*p0*np.exp(r*t)
    denominador = K+p0*(np.exp(r*t)-1)
    p = numerador/denominador
    return(p)
```

```python
# PROGRAMA ------
# INGRESO
p0 = 2
r  = 2
K  = 30
t  = 5

# PROCEDIMIENTO
ti = []
pi = []
i = 0
while i<=t:
    pi.append(f_logistica(p0,K,r,i))
    ti.append(i)
    i = i + 0.01

# SALIDA
np.set_printoptions(precision=3)
#print('tiempo:')
#print(ti)
#print('pi:')
#print(pi)

# grafica
plt.plot(ti,pi)
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.title('Solución de la ecuación de Verhulst')
plt.grid()
fname = "Verhlust_r_" + str(r)
plt.savefig(fname)
plt.show()
```
